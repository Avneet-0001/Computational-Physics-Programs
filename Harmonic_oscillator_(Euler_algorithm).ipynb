```python
import math
import matplotlib.pyplot as plt

tau = 0 # Dimensionless initial time
tau_final = 50 # Dimensionless final time
theta = math.pi/4 # Initial position
N = 10000 # Dividing total time into N smaller intervals
delta_tau = tau_final/N
v = 0 # Dimesionless inital velocity
time_list = [tau]
v_list = [v]
theta_list = [theta]

def f(theta,tau):
  return v

def g(theta,tau):
    return -math.sin(theta)

for i in range (0,N):
    theta_new = theta + delta_tau*f(theta,tau)
    v = v + delta_tau*g(theta,tau)
    theta = theta_new
    tau = tau + delta_tau
    theta_list.append(theta)
    v_list.append(v)
    time_list.append(tau)

plt.figure(figsize=(20,10))    
plt.xlabel('Time', fontweight='bold', fontsize='14')
plt.ylabel('Position , Velocity', fontweight='bold', fontsize='14')
plt.title('HARMONIC OSCILLATOR', fontweight='bold', fontsize='20')
plt.plot(time_list, theta_list, marker='D', label='x')
plt.plot(time_list, v_list, marker='^' , label='v')
plt.legend()
plt.show()     
```
